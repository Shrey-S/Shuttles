```python
def toStr(num):
    mm = int(num % 60)
    num = (num - mm) / 60
    if (mm < 10):
        mm = "0"+ str(mm)
    else:
        mm = str(mm)
    hh = int(num % 24)
    dd = int((num - hh) / 24)
    print("day %1d %2d:%2s" % (dd, hh, mm))
    return str(dd) + " " + str(hh) + ":" + str(mm)
def toStrArr(arr):
    newArr = []
    for x in arr:
        newArr.append(toStr(x))
def dif(str1, str2):
    return abs(toMin(str1) - toMin(str2))
```

```python
def toMin(mystr):
    mystr = mystr.replace("a","")
    mystr = mystr.replace("p","")
    mystr = mystr.replace("m","")
    arr = mystr.split(":")
    return int(int(arr[0]) * 60 + int(arr[1])) 

def toMinutes(mystr):
    if mystr == "12:00pm":
        return (12*60)
    mins = 0
    if mystr[-2] == "p":
        mystr = mystr.replace("p","")
        mins = mins + (12 * 60)
    mystr = mystr.replace("a","")
    mystr = mystr.replace("m","")
    arr = mystr.split(":")
    if arr[0] == "12":
        arr[0] == 0
    return mins + int(int(arr[0]) * 60 + int(arr[1])) 

def toMinSunday(mystr, day):        # minutes since Sunday 12 am 
    if mystr == "12:00pm":
        return (12*60)
    mins = 0
    if mystr[-2] == "p":
        mystr = mystr.replace("p","")
        mins = mins + (12 * 60)
    mystr = mystr.replace("a","")
    mystr = mystr.replace("m","")
    arr = mystr.split(":")
    hh = int(arr[0])
    mm = int(arr[1])
    if hh == 12:
        hh = 0
    mins = mins + hh * 60 + mm 
    if mins < (4 * 60):                  # if time is before 4 am
        mins = mins + (24 * 60)          # add 1 day in minutes
    mins = (mins + (day * 24 * 60)) % (7 * 24 * 60)  
    return mins
```
